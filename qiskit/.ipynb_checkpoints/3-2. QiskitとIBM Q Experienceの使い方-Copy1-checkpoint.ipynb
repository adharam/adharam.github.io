```python
#必要なモジュールのインポート
from qiskit import IBMQ, QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute, Aer
from qiskit.qasm import pi
from qiskit.tools.visualization import plot_histogram, circuit_drawer
import numpy as np
```

```python
# 自分のアカウント情報をloadする。（あらかじめ IBMQ.save_account を実行しておく必要がある. 複数のアカウントを使い分ける時はここで行う)
provider = IBMQ.load_account()

# 自分のアカウントで使用できるバックエンドを見る
provider.backends()
```

```python
backend_lb = provider.backends()[10]
backend_lb
```

```python

qreg_q = QuantumRegister(13, 'q')
creg_c = ClassicalRegister(13, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.h(qreg_q[3])
circuit.h(qreg_q[6])
circuit.cx(qreg_q[0], qreg_q[1])
circuit.cx(qreg_q[3], qreg_q[4])
circuit.cx(qreg_q[6], qreg_q[7])
circuit.cx(qreg_q[0], qreg_q[2])
circuit.cx(qreg_q[3], qreg_q[5])
circuit.cx(qreg_q[6], qreg_q[8])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.h(qreg_q[2])
circuit.h(qreg_q[3])
circuit.h(qreg_q[4])
circuit.h(qreg_q[5])
circuit.h(qreg_q[6])
circuit.h(qreg_q[7])
circuit.h(qreg_q[8])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5], qreg_q[6], qreg_q[7], qreg_q[8], qreg_q[9], qreg_q[10], qreg_q[11], qreg_q[12])
circuit.x(qreg_q[0])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5], qreg_q[6], qreg_q[7], qreg_q[8], qreg_q[9], qreg_q[10], qreg_q[11], qreg_q[12])
circuit.h(qreg_q[11])
circuit.h(qreg_q[12])
circuit.cx(qreg_q[11], qreg_q[0])
circuit.cx(qreg_q[11], qreg_q[1])
circuit.cx(qreg_q[11], qreg_q[3])
circuit.cx(qreg_q[11], qreg_q[4])
circuit.cx(qreg_q[11], qreg_q[6])
circuit.cx(qreg_q[11], qreg_q[7])
circuit.h(qreg_q[11])
circuit.cx(qreg_q[12], qreg_q[1])
circuit.cx(qreg_q[12], qreg_q[2])
circuit.cx(qreg_q[12], qreg_q[4])
circuit.cx(qreg_q[12], qreg_q[5])
circuit.cx(qreg_q[12], qreg_q[7])
circuit.cx(qreg_q[12], qreg_q[8])
circuit.h(qreg_q[12])
circuit.measure(qreg_q[11], creg_c[11])
circuit.measure(qreg_q[12], creg_c[12])
circuit.barrier(qreg_q[12], qreg_q[11], qreg_q[10], qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3], qreg_q[4], qreg_q[5], qreg_q[6], qreg_q[7], qreg_q[8], qreg_q[9])
circuit.cx(qreg_q[0], qreg_q[9])
circuit.cx(qreg_q[3], qreg_q[9])
circuit.cx(qreg_q[1], qreg_q[9])
circuit.cx(qreg_q[4], qreg_q[9])
circuit.cx(qreg_q[2], qreg_q[9])
circuit.cx(qreg_q[5], qreg_q[9])
circuit.cx(qreg_q[3], qreg_q[10])
circuit.cx(qreg_q[6], qreg_q[10])
circuit.cx(qreg_q[4], qreg_q[10])
circuit.cx(qreg_q[7], qreg_q[10])
circuit.cx(qreg_q[5], qreg_q[10])
circuit.cx(qreg_q[8], qreg_q[10])
circuit.measure(qreg_q[9], creg_c[9])
circuit.measure(qreg_q[10], creg_c[10])

#localのシミュレータとleast busyなbackend
backend_sim = Aer.get_backend("qasm_simulator")

#量子回路qcを指定したバックエンド(backend_sim)で4096回実行する。
result = execute(circuit, backend_sim, shots=1000).result()

#結果を出力する。
print(result.get_counts(circuit))

#結果のヒストグラムを描画する。
plot_histogram(result.get_counts(circuit))
```

```python
circuit.draw(output='mpl')
```

```python

#least busyだったbackendを選ぶ
backend_sim = backend_lb

#量子回路qcを指定したバックエンド(backend_sim)で4096回実行する。
result = execute(circuit, backend_sim, shots=4096).result()

#結果を出力する。
print(result.get_counts(circuit))

#結果のヒストグラムを描画する。
plot_histogram(result.get_counts(circuit))
```
